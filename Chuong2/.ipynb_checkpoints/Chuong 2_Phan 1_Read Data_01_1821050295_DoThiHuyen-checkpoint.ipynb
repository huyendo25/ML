**Yêu cầu 3.1:** Học viên tìm hiểu tập dữ liệu mẫu nhận dạng các loại rượu ( Wine recognition dataset).
1. Cho biết bộ dữ liệu này lưu trữ dữ liệu gì?
2. Có bao nhiêu thuộc tính? bao nhiêu bản ghi?
3. Đọc dữ liệu từ dataset mẫu bào biến dataframe.
4. Hiển thị một số mẫu rượu trong tập dữ liệu

```python
#{HỌC VIÊN VIẾT CODE Ở ĐÂY}
from sklearn.datasets import load_wine
wine = load_wine()
df_wine = pd.DataFrame(wine.data, columns=wine.feature_names) # Câu 3
df_wine["target"] = wine.target
#df_wine.info()
#https://scikit-learn.org/stable/datasets.html#datasets
print ('''Câu 1:
Bộ dữ liệu lưu trữ dữ liệu về rượu.
Dữ liệu là kết quả phân tích hóa học các loại rượu vang được trồng ở cùng một vùng ở Ý bởi ba người trồng trọt khác nhau.
Có mười ba phép đo khác nhau được thực hiện cho các thành phần khác nhau được tìm thấy trong ba loại rượu.\n''')
print ('''Câu 2:
Có 13 thuộc tính và 178 bản ghi.\n
Câu 3 và Câu 4 ở phần code có kèm cmt.''')
df_wine.tail () # Câu 4
```

**Yêu cầu 3.2:** Học viên tìm hiểu tập dữ liệu mẫu còn lại trong bộ Datasets của thư viện Sklearn:
1. Cho biết bộ dữ liệu này lưu trữ dữ liệu gì?
2. Có bao nhiêu thuộc tính? bao nhiêu bản ghi?
3. Tập dữ này dùng cho bài toán nào?

```python
#1.Tập dữ liệu Diabetes Dataset
#{Học viên mô tả các tập dữ liệu của thư viện Sklearn vào đây!}
from sklearn.datasets import load_diabetes
diabetes = load_diabetes()
df_diabetes = pd.DataFrame(diabetes.data, columns=diabetes.feature_names)
df_diabetes["target"] = diabetes.target
print ('''Câu 1:
Bộ dữ liệu này lưu trữ dữ liệu về bệnh tiểu đường.\n
Câu 2:
Số thuộc tính: 10
Số bản ghi: 442\n
Câu 3: Dùng cho bài toán hồi quy (regression)''')
df_diabetes
```

```python
#2.Tập dữ liệu Digits Dataset
#{Học viên mô tả các tập dữ liệu của thư viện Sklearn vào đây!}
from sklearn.datasets import load_digits
digits = load_digits()
df_digits = pd.DataFrame(digits.data, columns=digits.feature_names)
df_digits["target"] = digits.target
print ('''Câu 1:
Bộ dữ liệu này lưu trữ dữ liệu nhận dạng quang học của bộ dữ liệu chữ số viết tay.\n
Câu 2:
Số thuộc tính: 64
Số bản ghi: 1797\n
Câu 3: Dùng cho bài toán phân loại (classification)''')
df_digits
```

```python
#3.Tập dữ liệu Breast Cancer Wisconsin Dataset
#{Học viên mô tả các tập dữ liệu của thư viện Sklearn vào đây!}
from sklearn.datasets import load_breast_cancer
breast_cancer = load_breast_cancer()
df_BC = pd.DataFrame(breast_cancer.data, columns=breast_cancer.feature_names)
df_BC["target"] = breast_cancer.target
print ('''Câu 1:
Bộ dữ liệu này lưu trữ dữ liệu để chẩn đoán ung thư vú.\n
Câu 2:
Số thuộc tính: 30
Số bản ghi: 569\n
Câu 3: Dùng cho bài toán phân loại (classification)''')
df_BC
```
