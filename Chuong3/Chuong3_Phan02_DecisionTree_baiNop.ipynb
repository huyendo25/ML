## B| THỰC HÀNH TRÊN LỚP:

### B1. Đọc và chuẩn bị dữ liệu

```python
#Đọc tập dữ liệu đã xử lý trong chương 2
import pandas as pd
df_titanic = pd.read_csv('Data/Data_Titanic_ok.csv')

#Tách biến độc lập - biến phụ thuộc
#1.Tách biến độc lập X_titanic
X_titanic = df_titanic.iloc[:,1:].values

#2.Tách biến phụ thuộc y_titanic
y_titanic = df_titanic.iloc[:,0].values

#Tách thành tập Train-Test theo tỷ lệ 75-25:
from sklearn.model_selection import train_test_split
X_titanic_train, X_titanic_test, y_titanic_train, y_titanic_test = train_test_split (X_titanic,
                                                                                    y_titanic,
                                                                                    test_size=0.25,
                                                                                    random_state=42)
print ('1. Kích thước tập Training: ', X_titanic_train.shape)
print ('2. Kích thước tập Testing: ', X_titanic_test.shape)
```

### B.2 Xây dựng và đánh giá Model

```python
#Xây dựng model và huấn luyện:
from sklearn.tree import DecisionTreeClassifier


#Khai báo model với các tham số mặc định:
titanic_tree = DecisionTreeClassifier()


#Huấn luyện model với tập Train:
titanic_tree.fit(X_titanic_train, y_titanic_train)


#Độ chính xác của Model trên tập Huấn luyện
print("Độ chính xác của tập huấn luyện:",round(titanic_tree.score(X_titanic_train, y_titanic_train)*100,2))


#Đánh giá độ chính xác trên tập kiểm thử
y_titanic_pred = titanic_tree.predict(X_titanic_test)


from sklearn.metrics import accuracy_score
acc = accuracy_score(y_titanic_test, y_titanic_pred)
print("Độ chính xác trên tập kiểm thử đạt:", round(acc*100,2))


#Hiển thị confusion matrix:
import matplotlib.pyplot as plt
from sklearn.metrics import plot_confusion_matrix
plot_confusion_matrix(titanic_tree, X_titanic_test, y_titanic_test)  
plt.show()
```

#### Sử dụng grid search
---
Tham khảo GridSearchCV của thư viện sklearn tại <a href = 'https://scikit-learn.org/stable/modules/generated/sklearn.model_selection.GridSearchCV.html'>đây</a>

* **estimator** - (tham số đầu tiên) mô hình học máy Scikit-learning

* **param_grid** (dict) - từ điển tham số, dưới đây có 3 chiều và chứa 128 (8 x 4 x 4) kết hợp khác nhau, nghĩa là có thể đào tạo 128 mô hình khác nhau với Grid Search!

* **scoring** `default=None` - sử dụng để đo lường hiệu suất của mô hình. Ở đây dùng độ chính xác `accuracy`. Để xem đầy đủ, hãy nhấp vào <a href="https://scikit-learn.org/stable/modules/model_evaluation.html">đây</a>

* **n_jobs** `int, default=None` - số lượng công việc phải chạy song song `-1` có nghĩa là sử dụng tất cả các bộ xử lý (CPU).

* **cv** `int, default=None` - Số lần gấp để xác nhận chéo. Số tiêu chuẩn là `5`, `10`. Mỗi tổ hợp siêu tham số được lặp lại `10` lần vì `cv=10`. Vì vậy, tổng số lần lặp là 1280 (128 x 10).

* **return_train_score** `bool, default=False` - nếu `False` thì không bao gồm training score. Tính toán training score được sử dụng để hiểu rõ về cách các cài đặt thông số khác nhau ảnh hưởng đến sự cân bằng của việc overfitting/underfitting.

```python
#Xây dựng bộ các tham số
t_param_grid = {#'criterion':['entropy', 'gini'],
                #'splitter':['best', 'random'],
                #'max_features':[3,4,5,6],
                'max_depth':[3, 4, 6, 8, 10, 15, 20, 25], 
                'min_samples_leaf':[1, 2, 3, 4],
                'min_samples_split':[1, 2, 3, 4]}
                

#Sử dụng GridSearchCV để tự động hóa quá trình tối ưu hóa - điều chỉnh tham số
from sklearn.model_selection import GridSearchCV

titanic_tree1 = GridSearchCV(titanic_tree,
                             param_grid=t_param_grid, 
                             scoring="accuracy",
                             n_jobs=-1,
                             cv=10,
                             return_train_score=True)

titanic_tree1.fit(X_titanic_train, y_titanic_train)

#In ra bộ tham số tối ưu
print("Bộ tham số tối ưu:", titanic_tree1.best_params_)

#Độ chính xác của Model trên tập Huấn luyện
print("Độ chính xác của tập huấn luyện:",round(titanic_tree1.score(X_titanic_train, y_titanic_train)*100,2))

#Đánh giá độ chính xác trên tập kiểm thử
y_titanic_pred = titanic_tree1.predict(X_titanic_test)

acc = accuracy_score(y_titanic_test, y_titanic_pred)
print("Độ chính xác trên tập kiểm thử đạt:", round(acc*100,2)) #85,37
```

```python
#{HỌC VIÊN VIẾT CODE Ở ĐÂY}
#Thêm các thư viện
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn import tree
from sklearn.metrics import accuracy_score
import matplotlib.pyplot as plt
from sklearn.metrics import plot_confusion_matrix
from sklearn.model_selection import GridSearchCV
import numpy as np
import pandas as pd
```

## 1. Đọc dữ liệu mẫu từ Sklearn
---
* X_wine: chứa các feature đầu vào (input)
* y_wine: chứa label đầu ra (target)

```python
#Đọc dữ liệu mẫu Iris Dataset từ sklearn
from sklearn.datasets import load_wine
data = load_wine()
df_wine = pd.DataFrame(data['data'], columns=data['feature_names'])
df_wine['label'] = data['target']

#Đọc dữ liệu trả về biến array
X_wine, y_wine = load_wine(return_X_y=True)

print(type(X_wine))
print('Kích thước dữ liệu đầu vào (features):', X_wine.shape)
print('Kích thước dữ liệu đầu ra (target)   :', y_wine.shape)

#hiển thị số thực
np.set_printoptions(suppress=True)
print('Bộ dữ liệu 1)  ', X_wine[1,:], '--',y_wine[1])
print('Bộ dữ liệu 111) ', X_wine[111,:], '--',y_wine[111])
print('Bộ dữ liệu 155)', X_wine[155,:], '--',y_wine[155])
```

## 2. Tách tập dữ liệu ra tập Training - Test: 
---
* Training: dùng để huấn luyện mô hình (75%)
* Test: dùng để kiểm thử mô hình (25%)

```python
X_train, X_test, y_train, y_test = train_test_split(X_wine, y_wine, 
                                                    test_size=0.25, 
                                                    random_state=50)

print("1. Kích thước tập Training: %d" %len(y_train))
print("2. Kích thước tập Test    : %d" %len(y_test))
```

## 3. Sử dụng thuật toán Decision Tree để phân lớp
---
### 3.1 Sử dụng độ đo Entropy

```python
#Khai báo model Decision Tree sử dụng các tham số mặc định
wine_tree = DecisionTreeClassifier(criterion= 'entropy')

# Huấn luyên mô hình với dữ liệu huấn luyện
wine_tree.fit(X_train, y_train)

print("Độ chính xác của tập huấn luyện:",round(wine_tree.score(X_train, y_train)*100,2))
```

#### TRỰC QUAN HÓA CÂY QUYẾT ĐỊNH
---

```python
#Đặt tên cho các thuộc tính và các lớp rượu:
feature_names = list(data.feature_names)
target_names = list(data.target_names)
```

```python
#Cách 1: 
text_representation = tree.export_text(wine_tree,
                                       feature_names=feature_names,
                                       show_weights=True)
print(text_representation)
```

```python
#Cách 2:
fig = plt.figure(figsize=(25,20))

tree.plot_tree(wine_tree,
               feature_names=feature_names,
               class_names=target_names,
               filled=True)
plt.show()
```

```python
#Xác định mức độ quan trọng của các feature
print (feature_names)
print(wine_tree.feature_importances_)

#Trực quan hóa
plt.figure(figsize=(8,4))
plt.barh(feature_names,wine_tree.feature_importances_)

plt.title('MỨC ĐỘ QUAN TRỌNG CỦA CÁC THUỘC TÍNH')
plt.ylabel('Danh sách thuộc tính')
plt.xlabel('Độ quan trọng')
plt.grid(True)
plt.show()
```

```python
#Xác định độ sâu của cây: Khoảng cách tối đa giữa gốc và lá
print('Độ sâu của cây quyết định:',
      wine_tree.get_depth())

#Xác định số lượng nút lá của cây quyết định:
print('Số nút lá của cây quyết định:',
      wine_tree.get_n_leaves())
```

```python
#Đánh giá độ chính xác:
y_pred = wine_tree.predict(X_test)
acc = accuracy_score(y_test, y_pred)
acc_num = accuracy_score(y_test, y_pred, normalize=False)
print("Độ chính xác của mô hình trên tập test:", round(acc*100,2))
```

```python
#Hiển thị confusion matrix:
plot_confusion_matrix(wine_tree, X_test, y_test)  
plt.show()
```

### 3.2 Sử dụng độ đo Gini

```python
#Khai báo model Decision Tree sử dụng các tham số mặc định
wine_tree_g = DecisionTreeClassifier(criterion= 'gini')

# Huấn luyên mô hình với dữ liệu huấn luyện
wine_tree_g.fit(X_train, y_train)

print("Độ chính xác của tập huấn luyện:",round(wine_tree_g.score(X_train, y_train)*100,2))
```
